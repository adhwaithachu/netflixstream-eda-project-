```python
#Question1:chart showing common genres on netflix

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflixstream.csv')

genres_count = df['genres'].value_counts()

top_genres = genres_count.head(21)


plt.figure(figsize=(10,6))
top_genres.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title(" Common Genres on Netflix")
plt.xlabel("Genres")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Question2:Most common production company?
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflixstream.csv')

countries_count = df['production_countries'].value_counts()

top_countries = countries_count.head(31)

plt.figure(figsize=(10,6))
plt.plot(
    top_countries.index,
    top_countries.values,
    marker='o',
    linestyle='-',
    color='black'
)
plt.title("top Production Countries on Netflix")
plt.xlabel("Production Countries")
plt.ylabel("Number of Titles")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
#Question3:pie chart showing common production countries?
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('netflixstream.csv')

countries_count = df['production_countries'].value_counts()

top_countries = countries_count.head(11)

plt.figure(figsize=(8,8))
plt.pie(
    top_countries.values,
    labels=top_countries.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Paired.colors
)
plt.title("Top Production Countries on Netflix")
plt.tight_layout()
plt.show()
```

```python
#Question4:Which genre has the highest average IMDb score?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("netflixstream.csv")

sns.barplot(x='main_genre', y='imdb_score', data=df)
plt.title("Average IMDb Score by Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
#Question5:How many titles were released after the year 2015?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("netflixstream.csv")

# Filter titles released after 2015
recent = df[df['release_year'] > 2015]

sns.countplot(x='release_year', data=recent)
plt.title("Number of Titles Released After 2015")
plt.xlabel("Release Year")
plt.ylabel("Count of Titles")
plt.xticks(rotation=45)
plt.show()
```
